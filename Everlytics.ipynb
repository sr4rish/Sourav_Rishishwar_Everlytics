```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("test.json") as json_file:
    json_data = json.load(json_file)

# Print each key-value pair in json_data
for k in json_data.keys():
    print(k + ': ', json_data[k])
```

```python
fields = json_data['fields']
data = json_data['data']
```

```python
columns = []
for k in fields:
    columns.append(k['label'])
```

```python
dataset = pd.DataFrame(data)
dataset.columns = columns
dataset.head()
```

```python
dataset.describe()
```

```python
dataset_null = dataset.replace('NR', np.nan)
```

```python
dataset_num = dataset_null.apply(pd.to_numeric, errors='ignore')
```

```python
dataset_num.dtypes
```

```python
dataset_num['Upper Primary_Boys'] = pd.to_numeric(dataset_num['Upper Primary_Boys'], errors='coerce')
```

```python
dataset_num.dtypes
```

```python
dataset_num.head()
```

```python
dataset_num.isnull().sum()
```

```python
df1 = dataset_num[dataset_num.isna().any(axis=1)]
df1.head(10)
```

```python
dataset_num['State_UT'].unique()
```

```python
dataset_num = dataset_num.replace('Arunachal  Pradesh', 'Arunachal Pradesh')
states = dataset_num['State_UT'].unique()
states
```

```python
df1['Upper Primary_Boys'][0] != 'nan'
```

```python
#calculting mean drop out rate for each state across different features, not considering observation for this claculation
#with 'NR' in any one of section of students.
#for example A & N Islands	2013-14	1.35	1.06	1.21	NaN	1.54	0.51
#in this case, data for primary students was used for calculation, but data for upper primary students was not used for 
#mean drop out cal as primary_boys drop out rate was not reported.
data_new = []

for k in states:
    df = dataset_num[dataset_num['State_UT']==k]
    df = df.reset_index(drop=True)
    x_p = x_up = x_sec = x_hsec=0
    count_p = count_up = count_sec = count_hsec =0
    for i in range(len(df)):
        if(np.isnan(df['Primary_Total'][i])==False):
            
            if(np.isnan(df['Primary_Boys'][i])==False and np.isnan(df['Primary_Girls'][i])==False):
                x_p = df['Primary_Total'][i]+x_p
                count_p = count_p + 1
        
        if(np.isnan(df['Upper Primary_Total'][i])==False):
            
            if(np.isnan(df['Upper Primary_Boys'][i])==False and np.isnan(df['Upper Primary_Girls'][i])==False):
                x_up = df['Upper Primary_Total'][i]+x_up
                count_up = count_up + 1
        
        if(np.isnan(df['Secondary _Total'][i])==False):
            
            if(np.isnan(df['Secondary _Boys'][i])==False and np.isnan(df['Secondary _Girls'][i])==False):
                x_sec = df['Secondary _Total'][i]+x_sec
                count_sec = count_sec + 1
                
        if(np.isnan(df['HrSecondary_Total'][i])==False):
            
            if(np.isnan(df['HrSecondary_Boys'][i])==False and np.isnan(df['HrSecondary_Girls'][i])==False):
                x_hsec = df['HrSecondary_Total'][i]+x_hsec
                count_hsec = count_hsec + 1
        
        
    if(count_p != 0):
        mean_p = x_p/count_p
    else:
        mean_p = np.nan
    
    if(count_up != 0):
        mean_up = x_up/count_up
    else:
        mean_up = np.nan
        
    if(count_sec != 0):
        mean_sec = x_sec/count_sec
    else:
        mean_sec = np.nan
        
    if(count_hsec != 0):
        mean_hsec = x_hsec/count_hsec
    else:
        mean_hsec = np.nan
        
    data_new.append([k, mean_p, mean_up, mean_sec, mean_hsec])
    
```

```python
data_new[0:10]
```

```python
data_mean = pd.DataFrame(data_new)
```

```python
data_mean.columns = ['State', 'Mean_Primary_Dropout', 'Mean_Upper Primary_Dropout', 'Mean_Secondary_Dropout', 'Mean_HrSecondary_Dropout']
```

```python
data_mean.head()
```

```python
Total = []
for i in range(len(data_mean)):
    Total.append(np.mean(data_mean.iloc[i,1:]))
```

```python
data_mean['Mean_Dropout_State'] = Total
```
